```python
import matplotlib.pyplot as plt

damage_image=plt.imread('damage.jpg')
```

```python
import numpy as np
from skimage.restoration import inpaint

mask=get_mask(damage_image)

restored_image=inpaint.inpaint_biharmonic(damage_image,mask,multichannel=True)

plt.imshow(restored_image)
```

```python
def get_mask(damage_image):
    mask=np.zeros(damage_image.shape[:-1])
    mask[30:100,150:230]=1
    mask[100:300,120:170]=1
    return mask
```

```python
image=plt.imread('ronaldo.jpeg')
plt.imshow(image)
```

```python
from skimage.restoration import inpaint
mask=np.zeros(image.shape[:-1])
mask[700:800,1200:1350]=1
restored_image=inpaint.inpaint_biharmonic(image,mask,multichannel=True)
plt.imshow(restored_image,cmap='gray')
```
