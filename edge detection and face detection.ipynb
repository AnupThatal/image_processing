```python
import matplotlib.pyplot as plt
from skimage import color,data
from skimage.feature import canny

coins=data.coins()

coins_gray=color.rgb2gray(coins)

canny_image=canny(coins_gray,sigma=0.5)

plt.imshow(canny_image,cmap='gray')
```

```python
image=plt.imread('ronaldo.jpeg')

image_gray=color.rgb2gray(image)

plt.imshow(image_gray)
```

```python
from skimage.feature import Cascade
from skimage import data


train_file=data.lbp_frontal_face_cascade_filename()


detector=Cascade(train_file)

detected=detector.detect_multi_scale(img=image_gray,scale_factor=1.2,step_ratio=1,min_size=(10,10),max_size=(200,200))
```
