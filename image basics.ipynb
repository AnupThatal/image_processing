```python
#threshold is a technique where you seprate the object form image

from skimage import data,color

camera=data.camera()

camera_gray=color.rgb2gray(camera)

plt.imshow(camera)
```

```python
#for the threshold image

thresh=127

camera_1=camera_gray>127

camera_2=camera_gray<=127
```

```python
fig=plt.figure(figsize=(12,8))

ax=fig.add_subplot(1,3,1)
ax.imshow(camera_1)
plt.title('thershold image')


ax=fig.add_subplot(1,3,2)
ax.imshow(camera_2)
plt.title('thershold image')

ax=fig.add_subplot(1,3,3)
ax.imshow(camera)
plt.title('orginal image')
```

```python
#applying thershold_otsu 

from skimage.filters import threshold_otsu

from skimage import data,color

board=data.checkerboard()

board_gray=color.rgb2gray(board)

thresh=threshold_otsu(board_gray)

binary_image=board_gray>=thresh

fig=plt.figure(figsize=(12,8))

ax=fig.add_subplot(1,3,1)
ax.imshow(board_gray,cmap='gray')

ax=fig.add_subplot(1,3,2)
ax.imshow(binary_image,cmap='gray')
plt.title('binary_image')

ax=fig.add_subplot(1,3,3)
ax.imshow(board,cmap='gray')
```

```python
#threshold for image 
img=plt.imread('../deep learning/data/bookpage.jpg')
plt.imshow(img)
```

```python
from skimage import color
from skimage.filters import try_all_threshold

img_gray=color.rgb2gray(img)

image_thresh=try_all_threshold(img_gray,verbose=False)
```

```python
from skimage import color
from skimage.filters import threshold_otsu

img_gray=color.rgb2gray(img)

thresh=threshold_otsu(img_gray)

binary_image=thresh<img_gray

plt.imshow(binary_image)
```

```python
from skimage import color
from skimage.filters import threshold_local

img_gray=color.rgb2gray(img)

block_size=57

thresh=threshold_local(img_gray,block_size)

binary_image=thresh<img_gray

plt.imshow(binary_image,cmap='gray')
```
