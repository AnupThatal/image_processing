```python
from skimage.filters import sobel
from skimage import data,color
import matplotlib.pyplot as plt
```

```python
coins=data.coins()
coins=color.rgb2gray(coins)
image=sobel(coins)
plt.imshow(image,cmap='gray')
```

```python
from skimage.filters import gaussian
grass=data.grass()
grass_gray=color.rgb2gray(grass)

image1=gaussian(grass_gray)

plt.imshow(image1,cmap='gray')
```

```python
from skimage import data,color,exposure
import matplotlib.pyplot as plt
camera=data.camera()

camera_contrast=exposure.equalize_hist(camera)

plt.imshow(camera_contrast,cmap='gray')
```

```python
from skimage import data,color,exposure
import matplotlib.pyplot as plt
camera=data.camera()

camera_contrast=exposure.equalize_adapthist(camera,clip_limit=0.03)

plt.imshow(camera_contrast,cmap='gray')
```

```python
from skimage import data
from skimage.transform import rotate

rocket=data.rocket()

rocket_rotate=rotate(rocket,-90)

plt.imshow(rocket_rotate)
```

```python
from skimage import data
from skimage.transform import rescale

image=data.gravel()

img_rescale=rescale(image,1/3,anti_aliasing=True,multichannel=True)

plt.imshow(img_rescale)
```

```python
from skimage.transform import resize

height=int((image.shape[0])/3)
width=int(image.shape[1]/3)

image_resize=resize(image,(height,width),anti_aliasing=True)

plt.imshow(image_resize)
```

```python
from skimage import morphology

image=data.horse()

image_erosion=morphology.binary_erosion(image)

plt.imshow(image_erosion,cmap='gray')
```

```python
from skimage import morphology

image=data.horse()

image_dilation=morphology.dilation(image)

plt.imshow(image_dilation,cmap='gray')
```
